<h2> Kampala Rainfall of 2022 Data Cleaning</h2>

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Kampalanew.csv", encoding ="latin1")
df.head()
```

```python
df.shape
```

```python
df
```

```python
df_kampala =df.replace(['/24h','/12h','/6h','/3h','/1h'],['','','','',''], regex = True)
df_kampala.head()
```

```python
df_kampala.replace('Tr/24h', 0.01, inplace = True)
df_kampala.replace('Tr/12h', 0.01, inplace = True)
df_kampala.replace('Tr/6h', 0.01, inplace = True)
df_kampala.replace('Tr/3h', 0.01, inplace = True)
df_kampala.replace('Tr', 0.01, inplace = True)
```

```python
df_kampala.describe()
```

```python
# check for the missing data
missing_data = df_kampala.isnull()
missing_data.head()
```

```python
# we need to count the number of missing values
for column in missing_data.columns.values.tolist():
    print(column)
    print(missing_data[column].value_counts())
    print("")
```

<b> Missing colunms are TC, TdC, Hr, Nh, Precmm, Prechars</b>

```python
df_kampala.dtypes
```

```python
# change the dtype from object to float
df_kampala["Precmm"] = df_kampala["Precmm"].astype(float)
```

```python
df_kampala.dtypes
```

```python
# Precmm
ave_Precmm = df_kampala["Precmm"].mean()
print("The average of Precmm is:", ave_Precmm)
```

```python
#use method replace to fill the missing value in Precmm
df_kampala["Precmm"].replace(np.nan, ave_Precmm, inplace = True)
```

```python
#Prechars - get the mode 
df_kampala['Prechrs'].value_counts().idxmax()
```

```python
df_kampala["Prechrs"].replace(np.nan, "3h", inplace = True)
```

```python
df_kampala.dtypes
```

```python
# Compute the average for TC
ave_TC = df_kampala["TC"].mean()
print("The average of TC is:", ave_TC)
```

```python
# replace the missing values in TC with average
df_kampala["TC"].replace(np.nan, ave_TC, inplace = True)
```

```python
# Compute the average for TdC
ave_TdC = df_kampala["TdC"].mean()
print("The average of TdC is:", ave_TC)
```

```python
# replace the missing values in TdC with average
df_kampala["TdC"].replace(np.nan, ave_TdC, inplace = True)
```

```python
# Compute the average for Hr
ave_Hr = df_kampala["Hr"].mean()
print("The average of Hr is:", ave_Hr)
```

```python
# replace the missing values in Hr with average
df_kampala["Hr"].replace(np.nan, ave_Hr, inplace = True)
```

```python
# Compute the average for Nh
ave_Nh = df_kampala["Nh"].mean()
print("The average of Nh is:", ave_Nh)
```

```python
# replace the missing values in Nh with average
df_kampala["Nh"].replace(np.nan, ave_Nh, inplace = True)
```

```python
# You have to create a column called Rainfall and fill the records in it based on the following conditiond:
# For Precmm value 0 - 4  call it Light, 4 - 16 call it Mild_moderate, 16 -32 call it Moderate_heavy, 32 - 64 call it Heavy, 64 - 128 call it Heavy_torential, and above 128 call it Torential.
conditions = [
    (df_kampala["Precmm"] >= 0) & (df_kampala["Precmm"] < 4),
    (df_kampala["Precmm"] >= 4) & (df_kampala["Precmm"] < 16),
    (df_kampala["Precmm"] >= 16) & (df_kampala["Precmm"] < 32),
    (df_kampala["Precmm"] >= 32) & (df_kampala["Precmm"] < 64),
    (df_kampala["Precmm"] >= 64) & (df_kampala["Precmm"] < 128),
    (df_kampala["Precmm"] >= 128)
]
categories = ["Light", "Mild_moderate", "Moderate_heavy", "Heavy", "Heavy_torential", "Torential"]
df_kampala["Rainfall"] = pd.Series(np.select(conditions, categories))

df_kampala.head()
```
